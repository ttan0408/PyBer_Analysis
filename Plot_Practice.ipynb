```python
import numpy as np
# Import dependencies.
import matplotlib.pyplot as plt


```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.xlabel('Month')
plt.ylabel('US Dollars')
#plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="o", color="orange", linewidth=1.5, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt() Object-Oriented Function
# Define plot space
fig, ax = plt.subplots(figsize=(10, 6))

1-2-3# Define x and y axes
ax.plot(x_axis, 
        y_axis,
        marker = 'D',
        color = 'green',
        linewidth = 2 )

4 #Define legend
ax.legend(["Boston"])

#Add title
ax.set_title("PyBer Fare By Month")

#Add axis label
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#6 Add axis limit
ax.set_ylim(0,45)

#7 Add Grid Line
ax.grid()



```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="blue", label='Boston',linewidth = 2, edgecolor = "black",alpha=0.2, s=[i * 5 for i in y_axis])
ax.invert_yaxis()

#2 Define legend
ax.legend(["Boston"])

#3 Add title
ax.set_title("PyBer Fare By Month")

#4 Add axis label
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

#5 Set limit on x axis
ax.set_xlim(0, 50)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
